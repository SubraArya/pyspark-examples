```

import os
os.environ["JAVA_HOME"] = "/usr/lib/jvm/java-8-openjdk-amd64"
os.environ["SPARK_HOME"] = "/content/spark-2.4.5-bin-hadoop2.7"

import findspark
findspark.init()
from pyspark.sql import SparkSession
spark = SparkSession.builder.master("local[*]").getOrCreate()

print (os.listdir('./'))
df = spark.read.csv('rows_10000.csv',header=True)

df.show(5)


df.printSchema() 

df.columns 

df.head(5) 

df.select(['ID','Date']).show(5) 
df.groupBy('Location Description').count().orderBy('count',ascending=False).show(3)
from pyspark.sql.functions import min, max, date_add, upper, lower, substring ,col

df.select(upper(col('Location Description'))).show(2) 

df.select(lower(col('Location Description'))).show(2) 

df.select(substring(col('Location Description'),1,5)).show(2) 

df.select(max(col('Date'))).show(1) 

df.select(date_add(max(col('Date')),3)).show(1) 


ps = spark.read.csv('Police_Stations.csv',header=True)

from pyspark.sql.functions import lpad

ps = ps.withColumn('Padded_ID',lpad(col('DISTRICT'),3,'0'))

ps.show(5)

df.join(ps, df.District == ps.Padded_ID, 'left_outer').show(5)
```
